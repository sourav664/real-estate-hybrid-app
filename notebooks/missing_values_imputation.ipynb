```python
from sklearn.base import BaseEstimator, TransformerMixin

class FrequencyEncoder(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.freq_maps = {}

    def fit(self, X, y=None):
        df = X.copy()
        for col in df.columns:
            self.freq_maps[col] = df[col].value_counts(normalize=True).to_dict()
        return self

    def transform(self, X):
        df = X.copy()
        for col in df.columns:
            df[col] = df[col].map(self.freq_maps[col])
        return df
```

```python
freq_encod = FrequencyEncoder()
freq_encod.fit(df_age[['locality','projectname']])
```

```python
frequency_encoding = freq_encod.transform(df_age[['locality','projectname']])
```
