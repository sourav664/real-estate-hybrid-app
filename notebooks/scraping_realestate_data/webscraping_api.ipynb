```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
import os
from concurrent.futures import ThreadPoolExecutor, as_completed
import re

# -----------------------------
# CONFIGURATION
# -----------------------------
df = pd.read_csv("Data/delhiidsAndUrls.csv")  # your file
output_dir = "Data/propertyDetailsdelhi"
os.makedirs(output_dir, exist_ok=True)

headers = {
        'user-agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36"}

with open('error_ids.txt', 'r') as f:
    error_ids = f.read().splitlines()
# -----------------------------
# FUNCTION TO FETCH ONE PAGE
# -----------------------------
def fetch_property_details(property_id, property_url):
    try:
        url = f"https://www.magicbricks.com/propertyDetails/{property_url}"
        response = requests.get(url, headers=headers, timeout=15)

        if response.status_code != 200:
            print(f"❌ Failed {property_id} - HTTP {response.status_code}")
            return None

        html = response.text



        pattern = r'window\.SERVER_PRELOADED_STATE_DETAILS\s*=\s*(\{.*?\});'
        match = re.search(pattern, html, re.DOTALL)

        if match:
            json_text = match.group(1)
        else:
            with open("error_ids.txt", 'w') as f:
                f.write(f"{property_id}\n")  # Write the property ID to the fi
                
            
        output_file = os.path.join(output_dir, f"{property_id}.json")
        with open(output_file, "w", encoding="utf-8") as f:
            json.dump(json.loads(json_text), f, ensure_ascii=False, indent=4)
        print(f"\nSuccessfully saved data for property ID: {property_id}")
        return None
    except Exception as e:
        error_ids.append(property_id)
        print(f"❌ Error for {property_id}: {e}")
        return None

# -----------------------------
# MULTITHREADING FOR SPEED
# -----------------------------
results = []
with ThreadPoolExecutor(max_workers=20) as executor:
    futures = [
        executor.submit(fetch_property_details, row["id"], row["url"])
        for _, row in df.iterrows()
    ]
    for future in as_completed(futures):
        result = future.result()
        if result:
            results.append(result)

print(f"\n✅ Completed {len(results)} properties successfully.")
```
